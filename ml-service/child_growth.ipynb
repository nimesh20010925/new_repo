```python
IMAGE_ROOT = "/content/drive/MyDrive/child growth/images"
CSV_PATH = "/content/drive/MyDrive/child growth/labels.csv"
```

```python
import pandas as pd

df = pd.read_csv(CSV_PATH)

# Create health label
def health_status(row):
    if row['weight in grams'] < 12000 or row['waistline in cm'] < 48:
        return 0
    return 1

df['health'] = df.apply(health_status, axis=1)

# Create label dictionary
label_dict = {}

for _, row in df.iterrows():
    label_dict[row['child_id'].lower()] = {
        "gender": 0 if row['gender'] == 'f' else 1,
        "age": row['age in months'],
        "measurements": [
            row['height in cm'],
            row['weight in grams'],
            row['head circumference in cm'],
            row['waistline in cm']
        ],
        "health": row['health']
    }

print("✅ label_dict created with", len(label_dict), "children")
```

```python
import cv2
import numpy as np
import os

IMG_SIZE = 224

X_images, X_age = [], []
y_gender, y_measurements, y_health = [], [], []

for child_id in os.listdir(IMAGE_ROOT):

    if child_id.startswith('.'):
        continue

    child_id = child_id.lower()

    if child_id not in label_dict:
        continue

    child_path = os.path.join(IMAGE_ROOT, child_id)

    for img_name in os.listdir(child_path):
        img_path = os.path.join(child_path, img_name)

        img = cv2.imread(img_path)
        if img is None:
            continue

        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        img = img / 255.0

        X_images.append(img)
        X_age.append(label_dict[child_id]['age'])
        y_gender.append(label_dict[child_id]['gender'])
        y_measurements.append(label_dict[child_id]['measurements'])
        y_health.append(label_dict[child_id]['health'])

X_images = np.array(X_images)
X_age = np.array(X_age)
y_gender = np.array(y_gender)
y_measurements = np.array(y_measurements)
y_health = np.array(y_health)

print("✅ Total usable samples:", X_images.shape[0])
```

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split

# Split data into train and test (80-20)
X_train_img, X_test_img, X_train_age, X_test_age, y_train_gender, y_test_gender, y_train_measurements, y_test_measurements, y_train_health, y_test_health = train_test_split(
    X_images, X_age, y_gender, y_measurements, y_health, test_size=0.2, random_state=42
)

# Reshape age to (num_samples, 1)
X_train_age = X_train_age.reshape(-1, 1)
X_test_age  = X_test_age.reshape(-1, 1)

# Convert labels to float32
y_train_measurements = np.array(y_train_measurements, dtype=np.float32)
y_test_measurements  = np.array(y_test_measurements, dtype=np.float32)

y_train_gender = np.array(y_train_gender, dtype=np.float32)
y_test_gender  = np.array(y_test_gender, dtype=np.float32)

y_train_health = np.array(y_train_health, dtype=np.float32)
y_test_health  = np.array(y_test_health, dtype=np.float32)

print("✅ Data prepared for model")
```

```python
from tensorflow.keras.layers import Input, Dense, Conv2D, MaxPooling2D, Flatten, Dropout, concatenate
from tensorflow.keras.models import Model

# Image input
img_input = Input(shape=(IMG_SIZE, IMG_SIZE, 3))

x = Conv2D(32, (3,3), activation='relu')(img_input)
x = MaxPooling2D((2,2))(x)
x = Conv2D(64, (3,3), activation='relu')(x)
x = MaxPooling2D((2,2))(x)
x = Conv2D(128, (3,3), activation='relu')(x)
x = MaxPooling2D((2,2))(x)
x = Flatten()(x)
x = Dense(128, activation='relu')(x)

# Age input
age_input = Input(shape=(1,))
age_dense = Dense(32, activation='relu')(age_input)

# Combine image + age
combined = concatenate([x, age_dense])

# Shared dense layer
combined = Dense(128, activation='relu')(combined)
combined = Dropout(0.3)(combined)

# Outputs
height_output = Dense(1, name='height')(combined)
weight_output = Dense(1, name='weight')(combined)
head_output = Dense(1, name='head')(combined)
waist_output = Dense(1, name='waist')(combined)
gender_output = Dense(1, activation='sigmoid', name='gender')(combined)
health_output = Dense(1, activation='sigmoid', name='health')(combined)

# Model
model = Model(inputs=[img_input, age_input], outputs=[height_output, weight_output, head_output, waist_output, gender_output, health_output])

# Compile
model.compile(
    optimizer='adam',
    loss={
        'height': 'mse',
        'weight': 'mse',
        'head': 'mse',
        'waist': 'mse',
        'gender': 'binary_crossentropy',
        'health': 'binary_crossentropy'
    },
    metrics={
        'height': 'mae',
        'weight': 'mae',
        'head': 'mae',
        'waist': 'mae',
        'gender': 'accuracy',
        'health': 'accuracy'
    }
)

model.summary()
```

```python
history = model.fit(
    [X_train_img, X_train_age],
    [y_train_measurements[:,0], y_train_measurements[:,1], y_train_measurements[:,2], y_train_measurements[:,3], y_train_gender, y_train_health],
    validation_data=([X_test_img, X_test_age], [y_test_measurements[:,0], y_test_measurements[:,1], y_test_measurements[:,2], y_test_measurements[:,3], y_test_gender, y_test_health]),
    epochs=50,
    batch_size=16
)
```

```python
results = model.evaluate([X_test_img, X_test_age], [y_test_measurements[:,0], y_test_measurements[:,1], y_test_measurements[:,2], y_test_measurements[:,3], y_test_gender, y_test_health])
print(results)
```
